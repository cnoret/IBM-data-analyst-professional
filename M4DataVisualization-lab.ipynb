```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY, conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Query_1 = """
SELECT ConvertedComp
FROM master
"""

ConvComp = pd.read_sql_query(Query_1, conn)

sns.set_theme(style = "whitegrid")
plt.figure(figsize=(10,5))

sns.histplot(data = ConvComp["ConvertedComp"], 
             bins = 20, 
             edgecolor = "black",
             linewidth = 1.5)

plt.grid(True, 
         which = 'major', 
         linestyle = '--', 
         linewidth = 0.7, 
         color = 'gray', 
         alpha = 0.5)

plt.title("Distribution of Converted Compensation")
plt.xlabel("Converted Compensation")
plt.ylabel("Frequency")

plt.xticks(rotation = 45)

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
Query_2 = """
SELECT Age 
FROM master
"""

Age = pd.read_sql_query(Query_2, conn)

sns.set_theme(style="whitegrid")
plt.figure(figsize = (10, 5), facecolor = '#f9f9f9')

sns.boxplot(data = Age, x = "Age")

plt.title("Boxplot : Age", fontsize = 16, fontweight = 'bold', color = '#333')

plt.grid(True, 
         which = 'major', 
         linestyle = '--', 
         linewidth = 0.7, 
         color = 'gray', 
         alpha = 0.5)

plt.title("Box plot : Age")
plt.xlabel("Age", fontsize = 12, color = '#555')
plt.yticks([])

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
Query_3 ="""
SELECT Age, WorkWeekHrs
FROM master
"""

data = pd.read_sql_query(Query_3, conn)

plt.figure(figsize = (10, 6))
sns.set_theme(style = "whitegrid") 

sns.scatterplot(data = data, 
                x = "Age", 
                y = "WorkWeekHrs", 
                markers = "o",  
                s = 100,  
                alpha = 0.7,
                edgecolor = 'black')

plt.title("Relationship between Age and WorkWeekHrs", fontsize = 16, fontweight = 'bold')
plt.xlabel("Age", fontsize = 14)
plt.ylabel("WorkWeekHrs", fontsize = 14)

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
Query_4 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
data = pd.read_sql_query(Query_4, conn)

sns.set_theme(style="whitegrid")

plt.figure(figsize = (10, 6))

sns.scatterplot(data = data, 
                x = "WorkWeekHrs", 
                y = "CodeRevHrs", 
                size = "Age",  # Use Age for bubble size
                hue="Age",  # Optional: Use Age for color
                palette="viridis", 
                sizes = (20, 400),
                alpha = 0.6,
                edgecolor = 'black')

plt.title("Bubble Plot of Work Week Hours vs Code Review Hours", fontsize = 16, fontweight = "bold")
plt.xlabel("Work Week Hours", fontsize = 14)
plt.ylabel("Code Review Hours", fontsize = 14)
plt.legend(title = "Age", bbox_to_anchor = (1.05, 1), loc = "upper left")

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
Query_4 = """
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""

data = pd.read_sql_query(Query_4, conn)

x = data['count']
labels = data['DatabaseDesireNextYear']

plt.figure(figsize = (6, 6))

plt.pie(x, 
        labels = labels, 
        autopct = '%1.1f%%', 
        startangle = 90, 
        radius = 1.2)

plt.title("Top 5 Desired Databases for Next Year", fontsize = 16, fontweight = 'bold')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query_5 = """
SELECT AVG(WorkWeekHrs) AS AvgWorkWeekHrs, 
       AVG(CodeRevHrs) AS AvgCodeRevHrs, 
       Age 
FROM master 
WHERE Age BETWEEN 30 AND 35 
GROUP BY Age
"""

data = pd.read_sql_query(Query_5, conn)

data.set_index('Age', inplace = True)

plt.figure(figsize = (8, 6))
plt.bar(data.index, data['AvgWorkWeekHrs'], color = '#1f77b4', label = 'Work Week Hours')
plt.bar(data.index, data['AvgCodeRevHrs'], bottom = data['AvgWorkWeekHrs'], color = '#ff7f0e', label = 'Code Review Hours')

plt.ylabel("Hours", fontsize=14)
plt.xlabel("Age", fontsize=14)

plt.legend(loc = 'lower right')
plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
Query_6 = """
SELECT AVG(ConvertedComp) AS AvgConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP by Age
"""

data = pd.read_sql_query(Query_6, conn)

sns.set(style = "whitegrid")
plt.figure(figsize = (10, 6))

sns.lineplot(x = 'Age',
             y = 'AvgConvertedComp',
             data = data,
             marker = 'o')

# Add labels and title
plt.title("Median Converted Compensation by Age (45-60)", fontsize=16)
plt.xlabel("Age", fontsize = 14)
plt.ylabel("Converted Compensation", fontsize = 14)

# Show the plot
plt.tight_layout()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
Query_7 = """
SELECT MainBranch, count(*) AS count
FROM master
GROUP BY MainBranch
"""
data = pd.read_sql_query(Query_7, conn)

sns.set(style = "whitegrid")
plt.figure(figsize = (6, 4))

sns.barplot(x = 'count',
            y = 'MainBranch',
            data = data,
            edgecolor = "black"
            )

plt.grid(True, 
         which = 'major', 
         linestyle = '--', 
         linewidth = 0.7, 
         color = 'gray', 
         alpha = 0.5)

plt.title("Count of Main Branches", fontsize = 16)
plt.xlabel("Count", fontsize = 14)
plt.ylabel("")
plt.xticks(rotation = 45)

plt.show()
```
